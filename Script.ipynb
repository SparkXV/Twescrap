```python
import tweepy
import json
import sys
import csv
import urllib
import pandas as pd
from pandas import DataFrame
import jsonlines

#Twitter API credentials
consumer_key = "#########"
consumer_secret = "############"
access_key = "############"
access_secret = "##########"

tweets_images=[]
id_images=[]

def get_tweets(user_name):
 
    # authorization and initilization
    
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth)

    # list to store tweets
    tweets = []

    new_tweets = api.user_timeline(screen_name = user_name,count=199)

    tweets.extend(new_tweets)
    
    old_tweets = tweets[-1].id - 1

    while len(new_tweets) > 0:

        new_tweets = api.user_timeline(screen_name = user_name,count=199,max_id=old_tweets)

        tweets.extend(new_tweets)
        
        old_tweets = tweets[-1].id - 1

    print ("Total tweets downloaded from %s are %s" % (user_name,len(tweets)))
    
    return tweets


def store_tweets(alltweets):
    
    tweet_list=[]

    for tweet in alltweets:

        tweet_information=dict()

        # text of tweet
        tweet_information['text']=tweet.text

        # date and time at which tweet was created
        tweet_information['created_at']=tweet.created_at.strftime("%Y-%m-%d %H:%M:%S")

        # retweet count
        tweet_information['retweet_count']=tweet.retweet_count
    

        # favourites count
        tweet_information['favorite_count']=tweet.favorite_count

        tweet_list.append(tweet_information)
        
    with open('twitt_cred.json','w') as file_des:
        json.dump(tweet_list,file_des,indent=4,sort_keys=True)


        
if __name__ == '__main__':
    
    alltweets=get_tweets(input("Enter:"))                            # username as a input
    store_tweets(alltweets)
    
    
    
```
